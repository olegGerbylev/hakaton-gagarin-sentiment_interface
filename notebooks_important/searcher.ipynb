```python
nameAndSynonyms = pd.read_excel('names and synonyms_upd.xlsx')
```

```python
nameAndSynonyms.head()
```

```python
issueridAndSynonyms = {}
for i, n in nameAndSynonyms.iterrows():
  substingsArray = [x.strip() for x in n['Company_Ident_Text'].split(',') if x]
  issueridAndSynonyms.update({
      n['issuerid']: substingsArray
  })
```

```python
resultIssuerId = {'items': []}
# start_time = time.time()
for i, m in messageWithissuerid.iterrows():
  # if i == 2000:
  #   break;
  string = m['MessageText']
  for key, subStrings in issueridAndSynonyms.items():
      for subString in subStrings:
        if subString in string:
          resultIssuerId['items'].append({
              'issuerid': key,
              'synonym': subString,
              'messageid': m['messageid'],
              'ChannelID': m['ChannelID'],
              'DatePosted': m['DatePosted'],
              'MessageText': m['MessageText']
          })
          break
# end_time = time.time()
# execution_time = end_time - start_time
# print(f"–í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ü–∏–∫–ª–∞: {execution_time:.10f} —Å–µ–∫—É–Ω–¥")
df = pd.DataFrame(resultIssuerId['items'])
df['MessageText'][0]
```
