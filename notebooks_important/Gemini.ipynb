```python
import requests
import json
```

```python
def send_request(text, token):
    url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-pro:generateContent?key={token}"
    headers = {'Content-Type': 'application/json'}
    data = {"contents":[{"parts":[{"text": text}]}]}
    response = requests.post(url, headers=headers, data=json.dumps(data))
    return response.json()
```

# Разделение промта на два более простых

```python
def get_sentiment_from_text(company_name, text, gemini_api_token):
    # extraction sentences with company_name from text
    company_sentence_promt = f'Найди в тексте предложения, которые относятся к компании {company_name} и выведи только их. ОБЯЗАТЕЛЬНО НАЙДИ ВСЕ. Если упоминаний нет, выведи нет упоминаний. Текст: {text}'
    response = send_request(company_sentence_promt, gemini_api_token)
    company_sentences = response['candidates'][0]['content']['parts'][0]['text']

    print(f'Company Sentences: {company_sentences}')

    # retrieving sentiment from company sentences
    company_sentence_promt = f'Для следующих предложений определи общий сентимент в градации [Очень негативно, Скорее негативно, Нейтрально, Скорее позитивно, Очень позитивно] относительно компании: {company_name} и выведи только сентимент. Если компания не упоминается, выведи нет упоминаний. Предложения: {company_sentences}'
    response = send_request(company_sentence_promt, gemini_api_token)
    company_sentiment = response['candidates'][0]['content']['parts'][0]['text']

    return company_sentiment
```
